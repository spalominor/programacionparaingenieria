## Ejercicio 06

Seleccione una columna del DataFrame **original** que contenga una variable categórica (si no hay variables categóricas cree una columna con variables categóricas a partir de alguna columna numérica del DataFrame).

Utilice el método **groupby** para analizar la información del DataFrame a partir de la variable categórica seleccionada.

Muestre el resultado y las características del DataFrame.

```
# Escriba su código aquí
# Índices de las columnas de tipo 'against' en el dataframe
indices_against = np.arange(1, 19, 1)

# Tabla 2
# Obtener las columnas 'againist'
against_cols = df_pokemon.columns[indices_against]

# Usar groupby para analizar la información a partir de la columna categórica
df_06_against = df_pokemon.groupby('type1')[against_cols].mean()

# Mostrar el resultado
print("""Multiplicador promedio del daño recibido de cada tipo de Pokemón
        contra cada tipo de Pokemón\n""")
print(df_06_against)

# Mostrar las características del DataFrame
print(df_06_against.dtypes)
print("Forma del dataframe", df_06_against.shape)
print("\nTipos de Pokemón:", df_06_against.index)

# Tabla 1
# Marcar las columnas de interés
cols = ['attack', 'hp', 'defense', 'speed', 'base_happiness',
        'sp_attack', 'sp_defense', 'weight_kg', 'height_m']

# Usar groupby para analizar la información a partir de la columna categórica
df_06_stats = df_pokemon.groupby('type1')[cols].mean()

# Mostrar las características del dataframe
print(df_06_stats.dtypes)
print("Forma del dataframe:", df_06_stats.shape)
print("Tipos de Pokemón", df_06_stats.index)

# Mostrar el resultado
print("\n\nEstadísticas promedio de cada tipo de Pokemón")
print(df_06_stats)
```

Cree una tabla dinámica, utilizando el método **pivottable**, para analizar la información del DataFrame **original**.

Muestre el resultado y las características del DataFrame.

```
# Escriba su código aquí
# Determinar los índices de la tabla dinámica
indices = ['type1', 'type2', 'classfication']

# Determinar las columnas con los valores de la tabla dinamica
valores = against_cols

# Determinar las funciones de la tabla dinamica
funciones = ['mean']

# Imprimir la información necesaria para entender la tabla
print("""El valor es el multiplicador de daño recibido por el Pokemón
  2: El Pokémon es muy vulnerable al tipo de ataque específico.
  1: El Pokémon recibe daño normal del tipo de ataque específico.
  0.5: El Pokémon es resistente al tipo de ataque específico.
  0.25: El Pokémon es muy resistente al tipo de ataque específico.
  0: El Pokémon es inmune al tipo de ataque específico. No recibe ningún daño.
""")

# Construir la tabla dinámica sobre el dataframe original
td_pokemon_1 = df_pokemon.pivot_table(index=indices,
                                      values=valores,
                                      aggfunc=funciones)

# Imprimir el resultado
print(td_pokemon_1)
```

## Ejercicio 08

Encuentre todos los valores nulos del DataFrame original y cambie esos valores nulos con la media de la respectiva columna.

Muestre el resultado y las características del DataFrame.

```
# Escriba su código aquí
# Encontrar los valores nulos
nulos = df_pokemon.isnull().sum()

# Imprimir la cantidad de valores nulos
print(f"Valores nulos antes de la sustitución:\n{nulos}")

# Obtener las columnas numéricas
columnas_numericas = df_pokemon.select_dtypes(include='number').columns

# Calcular la media de cada columna numérica
medias = df_pokemon[columnas_numericas].mean()

# Llenar los valores nulos con las medias respectivas
df_pokemon.fillna(medias, inplace=True)

# Verificar si aún hay valores nulos
nulos_despues = df_pokemon.isnull().sum()

# Imprimir la cantidad de nulos después del cambio
print(f"\nValores nulos después de la sustitución:\n{nulos_despues}")

# Imprimir las características del DataFrame
print("\nCaracterísticas del DataFrame:\n")

# Imprimir la información del dataframe
print(df_pokemon.info())
```
