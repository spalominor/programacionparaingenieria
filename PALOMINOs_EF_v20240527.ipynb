## Preparación punto 00

* Importe las librerías requeridas

Cree el arreglo_00 de 15 **columnas** y 1000 **filas** de la siguiente manera:

* cinco columnas con números aleatorios provenientes de una distribución normal de media 0 y desviación típica 1. Utilice como semilla el número de su documento de identidad sumado con el número de la columna.

* cinco columnas con números aleatorios enteros en los siguientes intervalos: 200 a 300, 400 a 600, 0 a 360, 100 a 900, y 1000 a 2000. Utilice como semilla el número de su documento de identidad.

* cinco columnas con números aleatorios binarios (0,1)

```
# Escriba aquí su código

# Importar las librerías necesarias
import numpy as np



def crear_arreglo_00():
    """
    Crea un arreglo de 15 columnas y 1000 filas con números aleatorios
    según una semilla correspondiente al número de documento de identidad.

    Args:
    None

    Returns:
    numpy.ndarray: Arreglo de dimensiones (1000, 15)
    """
    # Establecer la semilla basada en el número de documento de identidad
    np.random.seed(1035972895)

    # Crear cinco columnas con números aleatorios de distribución normal
    normal_columns = np.random.normal(0, 1, size=(1000, 5))

    # Crear cinco columnas con números aleatorios enteros en los intervalos
    integer_columns = np.column_stack(
        [
            np.random.randint(200, 301, size=1000),
            np.random.randint(400, 601, size=1000),
            np.random.randint(0, 361, size=1000),
            np.random.randint(100, 901, size=1000),
            np.random.randint(1000, 2001, size=1000),
        ]
    )

    # Crear cinco columnas con números aleatorios binarios
    binary_columns = np.random.randint(0, 2, size=(1000, 5))

    # Combinar las tres partes en un solo arreglo
    arreglo_00 = np.column_stack([normal_columns,
                                  integer_columns,
                                  binary_columns])

    return arreglo_00


# Crear el arreglo_00 con la función anterior
arreglo_00 = crear_arreglo_00()

# Verificar la creación del arreglo
print(arreglo_00)
```

### gráfico 00

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de barras verticales. Coloque título al gráfico y etiquete los dos ejes.

```
# Escriba su código aquí

# Seleccionar las columnas pertinentes para analizar las estadísticas
columnas = ['hp', 'attack', 'defense', 'speed', 'sp_attack', 'sp_defense']

# Agrupar los datos por generación
# y calcular el promedio de todas las estadísticas
promedio_por_generacion = df_pokemon.groupby('generation')[columnas].mean()

# Obtener los índices de las generaciones (1, 2, ..., 7)
generaciones = promedio_por_generacion.index

# Obtener los promedios de todas las estadísticas
promedio_hp = promedio_por_generacion['hp']
promedio_attack = promedio_por_generacion['attack']
promedio_defense = promedio_por_generacion['defense']
promedio_speed = promedio_por_generacion['speed']
promedio_sp_attack = promedio_por_generacion['sp_attack']
promedio_sp_defense = promedio_por_generacion['sp_defense']

# Crear el gráfico de barras verticales
plt.figure(figsize=(10, 6))

# Dibujar las barras para cada estadística con su respectivo fondo
# Fondo 1: Puntos de vida
plt.bar(generaciones,
        promedio_hp,
        color='green',
        label='Puntos de Vida (HP)')

# Fondo 2: Ataque
plt.bar(generaciones,
        promedio_attack,
        color='skyblue',
        label='Ataque',
        bottom=promedio_hp)

# Fondo 3: Defensa
plt.bar(generaciones,
        promedio_defense,
        color='red',
        label='Defensa',
        bottom=promedio_hp + promedio_attack)

# Fondo 4: Velocidad
plt.bar(generaciones,
        promedio_speed,
        color='orange',
        label='Velocidad',
        bottom=promedio_hp + promedio_attack + promedio_defense)

# Fondo 5: Ataque especial
plt.bar(generaciones,
        promedio_sp_attack,
        color='purple',
        label='Ataque Especial',
        bottom=(promedio_hp
                + promedio_attack
                + promedio_defense
                + promedio_speed))

# Fondo 6: Defensa especial
plt.bar(generaciones,
        promedio_sp_defense,
        color='yellow',
        label='Defensa Especial',
        bottom=(promedio_hp
                + promedio_attack
                + promedio_defense
                + promedio_speed
                + promedio_sp_attack))


# Título y etiquetas de los ejes
plt.title('Promedio de Estadísticas por Generación')
plt.xlabel('Generación')
plt.ylabel('Puntos promedio')

# Establecer las etiquetas del eje x como las generaciones
plt.xticks(generaciones)

# Agregar leyenda y cuadrícula
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

### '01. Seleccionar datos

Ubique en un mapa los cinco puntos más alejados del centro del dataset.

```
# Escriba su código aquí

# Importar la clase Point para convertir coordenadas a objetos Point
from shapely.geometry import Point



def calcular_centro_dataset(gdf):
    """Calcula el centro del dataset utilizando la mediana de las coordenadas.

    Args:
        gdf (GeoDataFrame): El GeoDataFrame que contiene los datos.

    Returns:
        Point: El punto central del dataset.
    """
    # Proyectar los datos a un sistema de coordenadas proyectadas adecuado
    # Proyección de Mercator para evitar advertencias de Geopandas
    gdf = gdf.to_crs('EPSG:3395')

    # Calcular el centroide de cada país
    centroides = gdf.centroid

    # Calcular el punto medio de las latitudes y longitudes
    centro_latitud = centroides.y.mean()
    centro_longitud = centroides.x.mean()

    return Point(centro_longitud, centro_latitud)


def encontrar_puntos_mas_alejados(gdf, centro, n=100):
    """Encuentra los n puntos más alejados del centro del dataset.

    Args:
        gdf (GeoDataFrame): El GeoDataFrame que contiene los datos.
        centro (Point): El centro del dataset.
        n (int): El número de puntos a encontrar (por defecto, 100).

    Returns:
        GeoDataFrame: GeoDataFrame con los n puntos más alejados.
    """
    # Proyectar a un sistema de coordenadas proyectado
    # Proyección de Mercator para evitar advertencias de Geopandas
    gdf = gdf.to_crs("EPSG:3395")

    # Calcular la distancia de cada país al centro
    gdf['distancia_al_centro'] = gdf.distance(centro)

    # Retornar los n países más lejanos
    return gdf.nlargest(n, 'distancia_al_centro')


def dibujar_mapa_de_lejanos(gdf, outliers_gdf):
    """Plotea un mapa con todos los puntos y resalta los puntos más alejados.

    Args:
        gdf (GeoDataFrame): El GeoDataFrame que contiene todos los puntos.
        outliers_gdf (GeoDataFrame): El GeoDataFrame que contiene los puntos
        más alejados.

    Returns:
        None
    """
    # Proyectar el GeoDataFrame a un sistema de coordenadas geográficas
    gdf = gdf.to_crs('EPSG:4326')
    outliers_gdf = outliers_gdf.to_crs('EPSG:4326')

    # Crear el gráfico
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plotear los países más alejados
    gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
    outliers_gdf.plot(ax=ax, color='red', edgecolor='red')

    # Añadir título y etiquetas
    ax.set_title('Países Más Alejados del Centro'
                  ' del Dataset\n(Puntos en rojo)')
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')

    # Mostrar el gráfico
    plt.show()


# Calcular el centro del dataset
centro = calcular_centro_dataset(gdf)

# Encontrar los cinco puntos más alejados del centro
puntos_mas_alejados = encontrar_puntos_mas_alejados(gdf, centro, n=5)

# Imprimir los cinco puntos más alejados
print(puntos_mas_alejados['name'])

# Plotear un mapa con los puntos más alejados resaltados
dibujar_mapa_de_lejanos(gdf, puntos_mas_alejados)
```

### '00. Distancia de Manhattan

Calcule la distancia de Manhattan para los puntos del dataset. Luego indique el promedio de las distancias, la distancia mínima y la distancia máxima entre dos puntos. Muestre el par de coordenadas cuya distancia es mínima y el par de coordenadas cuya distancia es mínima.

```
# Escriba su código aquí

# Importar el módulo "distance" de SciPy spatial
from scipy.spatial import distance



def calcular_distancias_manhattan(gdf):
    """Calcula la distancia de Manhattan entre los puntos del dataset.

    Args:
        gdf (GeoDataFrame): El GeoDataFrame que contiene los datos.

    Returns:
        DataFrame: Un DataFrame con las distancias de Manhattan
        entre los puntos
    """
    # Proyectar los datos a un sistema de coordenadas proyectadas adecuado
    gdf_projected = gdf.to_crs('EPSG:3395')

    # Obtener los centroides de los países
    centroides = gdf_projected.geometry.centroid

    # Crear una matriz de coordenadas de los centroides
    coords = list(zip(centroides.x, centroides.y))

    # Calcular la distancia de Manhattan entre los puntos utilizando SciPy
    dist_matrix = distance.cdist(coords, coords, metric='cityblock')

    # Convertir la matriz de distancias a un DataFrame
    dist_df = pd.DataFrame(dist_matrix, index=gdf['name'],
                           columns=gdf['name'])

    return dist_df


def mostrar_resultados(distancias):
    """
    Función que muestra información importante de una matriz de distancias.
    Sirve para ahorrar líneas de código. La información que imprime es:

      - Promedio de las distancias
      - Distancia mínima y los puntos que la generan
      - Distancia máxima y los puntos que la generan

    Args:
      distancias (pd.DataFrame): Matriz de distancias

    Returns:
      None
    """
    # Calcular el promedio de las distancias
    promedio_distancia = distancias.mean().mean()

    # Encontrar la distancia mínima y máxima, excluyendo la diagonal
    distancia_minima = distancias[distancias > 0].min().min()
    distancia_maxima = distancias.max().max()

    # Encontrar los pares de coordenadas para la distancia mínima y máxima
    min_coords = distancias[distancias == distancia_minima].stack().index[0]
    max_coords = distancias[distancias == distancia_maxima].stack().index[0]

    # Mostrar los resultados
    print(f"Promedio de las distancias: {promedio_distancia} metros")
    print(f"Distancia mínima: {distancia_minima} metros entre {min_coords}")
    print(f"Distancia máxima: {distancia_maxima} metros entre {max_coords}")


# Calcular las distancias de Manhattan con la función
distancias_manhattan = calcular_distancias_manhattan(gdf)

# Mostrar el inicio del DataFrame con los resultados
print(distancias_manhattan.head())

# Mostrar los resultados solicitados sobre la información con estas distancias
mostrar_resultados(distancias_manhattan)
```

### '02. Triangulación de Delaunay

Aplique la [Triangulación de Delaunay](https://en.wikipedia.org/wiki/Delaunay_triangulation) para los puntos que se encuentran en el dataset y **muestre** el gráfico correspondiente.

```
# Escriba su código aquí

# Importar la clase Delaunay de Scipy spatial
from scipy.spatial import Delaunay



def aplicar_triangulacion_delaunay(gdf):
    """Aplica la Triangulación de Delaunay a los puntos del GeoDataFrame.

    Args:
        gdf (GeoDataFrame): El GeoDataFrame que contiene los datos.

    Returns:
        array: Los índices de los vértices de los triángulos.
        array: Las coordenas de los centros de cada país
    """
    # Proyectar los datos a un sistema de coordenadas proyectadas adecuado
    gdf_projected = gdf.to_crs('EPSG:3395')

    #EPSG:4326 es una proyección más natural a la vista, pero causa WARNINGS
    #gdf_projected = gdf.to_crs('EPSG:4326')

    # Obtener los centroides de los países proyectados
    centroides = gdf_projected.geometry.centroid

    # Crear una matriz de coordenadas de los centroides
    coords = list(zip(centroides.x, centroides.y))

    # Aplicar la Triangulación de Delaunay
    delaunay = Delaunay(coords)

    return delaunay, centroides


# Obtener la triangulación y los centroides para dibujar
triangulacion, centroides = aplicar_triangulacion_delaunay(gdf)

# Crear el gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Plotear los países
gdf.plot(ax=ax, color='lightgrey', edgecolor='black')

# Plotear la triangulación con líneas azules encima de los países
plt.triplot(centroides.x,
            centroides.y,
            triangulacion.simplices,
            color='blue')

# Añadir título y etiquetas
ax.set_title('Mapamundi con Triangulación de Delaunay')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')

# Mostrar el gráfico
plt.show()
```
