## Ejercicio 00

Cree el arreglo_00 de 15 **columnas** y 1000 **filas** de la siguiente manera:

* cinco columnas con números aleatorios provenientes de una distribución normal de media 0 y desviación típica 1. Utilice como semilla el número de su documento de identidad sumado con el número de la columna.

* cinco columnas con números aleatorios enteros en los siguientes intervalos: 200 a 300, 400 a 600, 0 a 360, 100 a 900, y 1000 a 2000. Utilice como semilla el número de su documento de identidad.

* cinco columnas con números aleatorios binarios (0,1)

Muestre el resultado y las características del arreglo.

```
# Escriba aquí su código
# Importar librerias
import numpy as np


def crear_arreglo_00():
    """
    Crea un arreglo de 15 columnas y 1000 filas con números aleatorios
    según una semilla correspondiente al número de identificación.

    Args:
    None

    Returns:
    numpy.ndarray: Arreglo de dimensiones (1000, 15) con los números aleatorios.
    """
    # Establecer la semilla basada en el número de documento de identidad
    np.random.seed(1035972895)

    # Crear cinco columnas con números aleatorios de distribución normal
    normal_columns = np.random.normal(0, 1, size=(1000, 5))

    # Crear cinco columnas con números aleatorios enteros en los intervalos
    integer_columns = np.column_stack(
        [
            np.random.randint(200, 301, size=1000),
            np.random.randint(400, 601, size=1000),
            np.random.randint(0, 361, size=1000),
            np.random.randint(100, 901, size=1000),
            np.random.randint(1000, 2001, size=1000),
        ]
    )

    # Crear cinco columnas con números aleatorios binarios
    binary_columns = np.random.randint(0, 2, size=(1000, 5))

    # Combinar las tres partes en un solo arreglo
    arreglo_00 = np.column_stack([normal_columns, integer_columns, binary_columns])

    return arreglo_00


# Crear el arreglo
arreglo_00 = crear_arreglo_00()

# Mostrar el resultado y las características del arreglo
print("Arreglo creado:")
print(arreglo_00)
print("\nDimensiones del arreglo:", arreglo_00.shape)
print("Tipo de datos del arreglo:", arreglo_00.dtype)
print("Valor máximo en el arreglo por columna:\n", np.max(arreglo_00, axis=0))
print("Valor mínimo en el arreglo por columna:\n", np.min(arreglo_00, axis=0))
print("Promedio de los elementos del arreglo:", np.mean(arreglo_00))
```

```
# Escriba aquí su código
# Importar librerías necesarias
import numpy as np


def crear_arreglo_06():
    """
    Crea un nuevo arreglo añadiendo nuevas columnas a este que contienen la
    suma de cada fila y el promedio de las nueve columnas originales.

    Args:
    None

    Returns:
    numpy.ndarray: Arreglo con las especificaciones.
    """
    # Crear un nuevo arreglo de 9x9 a partir del arreglo_00
    arreglo_temporal = arreglo_00[:9, :9]

    # Calcular la suma de cada fila
    suma_filas = np.sum(arreglo_temporal, axis=1)

    # Calcular el promedio de las nueve columnas originales
    promedio_columnas = np.mean(arreglo_temporal, axis=0)

    # Unir las nuevas columnas como un arreglo
    nuevas_columnas = np.column_stack([suma_filas, promedio_columnas])

    # Crear el arreglo_06 y agregar las 2 nuevas columnas al arreglo
    arreglo_06 = np.hstack((arreglo_temporal, nuevas_columnas))

    return arreglo_06


# Crear el arreglo solicitado
arreglo_06 = crear_arreglo_06()

# Mostrar el resultado y las características del arreglo resultante
print("Arreglo obtenido:")
print(arreglo_06)
print("\nDimensiones del arreglo obtenido:", arreglo_06.shape)
print("Tipo de datos del arreglo obtenido:", arreglo_06.dtype)
print("Valor máximo en el arreglo obtenido:", np.max(arreglo_06))
print("Valor mínimo en el arreglo obtenido:", np.min(arreglo_06))
print("Promedio de los elementos del arreglo obtenido:", np.mean(arreglo_06))
```

## Ejercicio 08

Cree un arreglo, de 15 columnas y 15 filas seleccionadas aleatoriamente (utilice el número de su documento de identidad como semilla), a partir del arreglo_00.

Calcule y muestre el determinante del arreglo, la transpuesta y la inversa.

```
# Escriba aquí su código
# Importar librerias necesarias
import numpy as np


def crear_arreglo_08():
    """
    Crea un arreglo 15x15 seleccionado aleatoriamente del arreglo_00
    con la semilla del número de identificación

    Args:
    None

    Returns:
    Un arreglo de numpy según lo especificado
    """
    # Establecer la semilla basada en el número de documento de identidad
    np.random.seed(1035972895)

    # Seleccionar el indice aleatorio para las filas a seleccionar
    filas_seleccionadas = np.random.choice(1000 - 15)

    # Crear el arreglo 15x15 según las filas seleccionadas
    arreglo_08 = arreglo_00[filas_seleccionadas : filas_seleccionadas + 15, :]

    # Retornar el arreglo
    return arreglo_08


# Crear el arreglo solicitado
arreglo_08 = crear_arreglo_08()

# Imprimir el arreglo
print("Arreglo obtenido:")
print(arreglo_08)

# Calcular e imprimir la transpuesta
print("Transpuesta:")
print(arreglo_08.T)

# Calcular e imprimir el determinante
print("Determinante:")
print(np.linalg.det(arreglo_08))

# Calcular e imprimir la inversa
print("Inversa:")
print(np.linalg.inv(arreglo_08))
```

## Ejercicio 09

Cree un arreglo, de 8 columnas y  8 filas, ambas seleccionadas aleatoriamente (utilice el número de su documento de identidad como semilla), a partir del arreglo_00.

Luego cree un arreglo de dimensión 1x8 con número reales aleatorios en el intervalo 0,10 (utilice el número de su documento de identidad como semilla).

Asuma que el primer arreglo representa los coeficientes de un sistema de ecuaciones lineales, y el segundo arreglo contiene las respectivas constantes. Encuentre la solución al sistema utilizando funciones de álgebra lineal de NumPy.

Muestre el valor de cada una de las variables que resuelve el sistema de ecuaciones.

```
# Escriba aquí su código
# Importar librerias necesarias
import numpy as np


def crear_arreglo_09():
    """
    Crea un arreglo de 8x8 seleccionado aleatoriamente del arreglo_00
    con la semilla del número de identificación, que es un arreglo de
    coeficientes. También crea un arreglo de constantes con numeros reales
    de 0 a 10 de tamaño 1x8

    Args:
    None

    Returns:
    numpy.ndarray: Arreglo de coeficientes del sistema de ecuaciones.
    numpy.ndarray: Arreglo de constantes del sistema de ecuaciones.
    """
    # Establecer la semilla basada en el número de documento de identidad
    np.random.seed(1035972895)

    # Seleccionar aleatoriamente 8 filas del arreglo_00
    filas_selec = np.random.choice(arreglo_00.shape[0], 8, replace=False)
    # Seleccionar aleatoriamente 8 columnas del arreglo_00
    columnas_selec = np.random.choice(arreglo_00.shape[1], 8, replace=False)

    # Seleccionar las entradas correspondientes a las filas y columnas
    arreglo_coeficientes = arreglo_00[filas_selec][:, columnas_selec]

    # Crear un arreglo de dimensiones 1x8 con números reales aleatorios
    arreglo_constantes = np.random.uniform(0, 10, size=(8, 1))

    # Retornar los arreglos solicitados
    return arreglo_coeficientes, arreglo_constantes


# Crear la matriz de coeficientes y de constantes
coeficientes, constantes = crear_arreglo_09()

# Obtener el vector solución x: Ax=b
solucion = np.linalg.solve(coeficientes, constantes)

# Imprimir el valor solución de cada una de las variables
print(solucion)
```
