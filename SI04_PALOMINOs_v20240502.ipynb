## Ejercicio 04

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de barras horizontales. Coloque título al gráfico y etiquete los dos ejes.

```
# Escriba su código aquí

# Seleccionar las columnas pertinentes del DataFrame
columnas_pertinentes = ['base_happiness',
                        'base_egg_steps',
                        'base_total']

# Agrupar los datos por generación y calcular el promedio de cada columna
promedios_generacion = df_pokemon.groupby('generation')[
    columnas_pertinentes].mean()

# Crear el gráfico de barras horizontales apiladas
plt.figure(figsize=(10, 6))

# Definir los colores para las barras
colores = ['orange', 'purple', 'green']

# Crear la primera barra
barra_base_happiness = plt.barh(promedios_generacion.index,
                                promedios_generacion['base_happiness'],
                                color=colores[0],
                                label='Base happiness')

# Crear la segunda barra empezando donde termina la primera
barra_base_egg_steps = plt.barh(promedios_generacion.index,
                                promedios_generacion['base_egg_steps'],
                                color=colores[1],
                                label='Base egg steps',
                                left=promedios_generacion['base_happiness'])

# Calcular las posiciones para la tercera barra
posiciones_base_total = promedios_generacion[
    'base_happiness'] + promedios_generacion['base_egg_steps']

# Crear la tercera barra empezando donde termina la segunda
barra_base_total = plt.barh(promedios_generacion.index,
                            promedios_generacion['base_total'],
                            color=colores[2],
                            label='Base total',
                            left=posiciones_base_total)

# Título y etiquetas de los ejes
plt.title('Promedio de las estadísticas base de los Pokémon por Generación')
plt.xlabel('Valor promedio')
plt.ylabel('Generación')

# Agregar leyenda y la cuadrícula
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

## Ejercicio 05

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de barras verticales. Coloque título al gráfico y etiquete los dos ejes.

```
import matplotlib.pyplot as plt

# Agrupar los datos por generación y calcular el promedio de 'hp' y 'attack'
promedio_por_generacion = df_pokemon.groupby('generation')[['hp',
                                                            'attack']].mean()

# Obtener los índices de las generaciones (1, 2, ..., 7)
generaciones = promedio_por_generacion.index

# Obtener los promedios de puntos de vida y ataque
promedio_hp = promedio_por_generacion['hp']
promedio_attack = promedio_por_generacion['attack']

# Crear el gráfico de barras verticales
plt.figure(figsize=(10, 6))

# Dibujar la barra según los puntos de vida
plt.bar(generaciones,
        promedio_hp,
        color='green',
        label='Puntos de Vida (HP)')

# Dibujar la barra según el ataque encima de la barra de puntos de vida
plt.bar(generaciones,
        promedio_attack,
        color='skyblue',
        label='Ataque',
        bottom=promedio_hp)

# Título y etiquetas de los ejes
plt.title('Promedio de Puntos de Vida y Ataque por Generación')
plt.xlabel('Generación')
plt.ylabel('Promedio')

# Establecer las etiquetas del eje x como las generaciones
plt.xticks(generaciones)

# Agregar leyenda y cuadrícula
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

## Ejercicio 07

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de torta (**pie chart**). Coloque título al gráfico y etiquete los segmentos del gráfico.

```
# Escriba su código aquí

# Seleccionar las columnas pertinentes del DataFrame
columna_pertinente = 'type1'

# Contar la cantidad de Pokemónes por su tipo
conteo_valores = df_pokemon[columna_pertinente].value_counts()

# Obtener los valores y las etiquetas para el gráfico de torta
valores = conteo_valores.values
etiquetas = conteo_valores.index

# Colores personalizados (usando una paleta de colores predefinida)
colores = plt.cm.tab20c.colors[:len(etiquetas)]

# Creamos una lista explode que tenga un valor de 0 para todos los
# segmentos excepto los últimos cinco.
# Esto resalta los últimos 5, los más raros
explode = np.zeros(len(etiquetas))
explode[-5:] = 0.1

# Crear el gráfico de torta, con los colores y los resaltos
plt.figure(figsize=(8, 8))
plt.pie(valores,
        labels=etiquetas,
        colors=colores,
        explode=explode,
        autopct='%1.1f%%',
        shadow=False
        )

# Título del gráfico
plt.title('Distribución de Pokémon por Tipo')

# Normalizar el ancho y alto de la figura
plt.axis('equal')

# Mostrar el gráfico
plt.show()
```

## Ejercicio 08

Seleccione las columnas que considere pertinentes del dataset y haga un mapa de calor (**heatmap**). Coloque título al gráfico y etiquete los dos ejes.

```
# Escriba su código aquí

# Seleccionar las columnas pertinentes del DataFrame
columnas_pertinentes = ['generation',
                        'hp',
                        'attack',
                        'defense',
                        'speed',
                        'sp_attack',
                        'sp_defense']

# Filtrar el DataFrame para incluir solo las columnas pertinentes
datos_seleccionados = df_pokemon[columnas_pertinentes]

# Agrupar los datos por generación y calcular el promedio de cada columna
promedios_generacion = datos_seleccionados.groupby('generation').mean()

# Crear un mapa de calor (heatmap) de los promedios por generación
plt.figure(figsize=(10, 6))
heatmap = plt.imshow(promedios_generacion.T,
                     cmap='viridis',
                     aspect='auto',
                     interpolation='nearest')

# Añadir título y etiquetas de los ejes
plt.title('Promedios de Atributos por Generación')
plt.xlabel('Generación')
plt.ylabel('Atributo')

# Añadir barra de color
plt.colorbar(heatmap, label='Valor Promedio')

# Ajustar la ubicación de las etiquetas del eje x
plt.xticks(np.arange(len(promedios_generacion.index)),
           promedios_generacion.index)

# Obtener los nombres de los atributos
nombres_atributos = ['HP',
                     'Ataque',
                     'Defensa',
                     'Velocidad',
                     'Super Ataque',
                     'Super Defensa']

# Añadir nombres de los atributos a la barra de color
plt.yticks(np.arange(len(nombres_atributos)), nombres_atributos)

# Mostrar el mapa de calor
plt.show()
```
