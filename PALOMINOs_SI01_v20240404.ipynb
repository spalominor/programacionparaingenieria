## Ejercicio 00

Escriba un programa que lea el archivo de referencia y reporte cuántas direcciones de correo electrónico contiene en total y cuantas por cada dominio de correo electrónico.




```python
# Resuelva en esta casilla el ejercicio 00
import pandas as pd
import re

# Guardar en la variable 'ruta' la url del dataset
ruta_csv = (
    "https://docs.google.com/spreadsheets/d/e/"
    "2PACX-1vTRfE6-FxXXP9jFLrFCuQZr6MTB5uYToBnJ"
    "QJKqjuQJfESFAsh5-fC9A14lpLfFnda3if8ioCkbhsfW/"
    "pub?gid=1040631360&single=true&output=csv"
)

# Cargar el dataset a partir de la ruta establecida
try:
    df = pd.read_csv(ruta_csv, header=None, dtype=str)
    df = df.rename(columns={0: "texto"})
    print("Lectura del dataset exitosa")
except Exception as e:
    print(f"Error al leer el dataset: {e}")

# Función para contar los correos
def contar_correos(dataset):
    """
    Cuenta la cantidad de direcciones de correo electrónico en un DataFrame.

    Args:
    - dataset: Un DataFrame de pandas que texto de todo tipo.

    Returns:
    - pandas.Series: Una serie de pandas que contiene los correos.
    """
    patron_correo = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    # Construir una lista con los correos encontrados
    lista_direcciones_correo = [
        direccion_correo
        for fila in dataset["texto"]
        for direccion_correo in re.findall(patron_correo, fila)
    ]
    # Convertir la lista en una Serie de pandas
    serie_direcciones_correo = pd.Series(lista_direcciones_correo)
    return serie_direcciones_correo


def contar_por_dominio(serie_direcciones):
    """
    Cuenta la cantidad de direcciones de correo electrónico para cada dominio.

    Args:
    - serie_direcciones: Una Serie que contiene las direcciones de correo.

    Returns:
    - pandas df: Un DataFrame con los dominios y la cantidad para cada uno.
    """
    # Diccionario para almacenar la cantidad de direcciones por dominio
    cantidad_dominio = {}

    # Iterar sobre cada dirección de correo electrónico en la Serie
    for direccion in serie_direcciones:
        # Extraer el dominio de la dirección de correo electrónico
        dominio = direccion.split('@')[-1]
        # Incrementar el contador para el dominio en el diccionario
        cantidad_dominio[dominio] = cantidad_dominio.get(dominio, 0) + 1

    df_dominio_cantidad = pd.DataFrame(
        list(cantidad_dominio.items()),
        columns=['Dominio', 'Cantidad']
    )

    return df_dominio_cantidad


# Contar la cantidad total de direcciones de correo electrónico en el DataFrame
total_correos = contar_correos(df)
# Obtener la cantidad de direcciones por dominio
cantidad_por_dominio = contar_por_dominio(total_correos)


# Imprimir el resultado
print(f"Total de correos electrónicos en el DataSet: {len(total_correos)}")
print(cantidad_por_dominio)
```

```python
# Resuelva en esta casilla el ejercicio 02
import pandas as pd
import re

# Guardar en la variable 'ruta' la url del dataset
ruta_csv = (
    "https://docs.google.com/spreadsheets/d/e/"
    "2PACX-1vTRfE6-FxXXP9jFLrFCuQZr6MTB5uYToBnJ"
    "QJKqjuQJfESFAsh5-fC9A14lpLfFnda3if8ioCkbhsfW/"
    "pub?gid=1040631360&single=true&output=csv"
)

# Cargar el dataset a partir de la ruta establecida
try:
    df = pd.read_csv(ruta_csv, header=None, dtype=str)
    df = df.rename(columns={0: "texto"})
    print("Lectura del dataset exitosa")
except Exception as e:
    print(f"Error al leer el dataset: {e}")

# Función para extraer fechas válidas
def extraer_fechas(dataset, format_default=True):
    """
    Extrae las fechas válidas en formato "DD/MM/AAAA" de un DataFrame.

    Args:
    - dataset: Un DataFrame de pandas que contiene texto de todo tipo.

    Returns:
    - pandas.Series: Una serie que contiene las fechas válidas.
    """
    # Formato DD/MM/AAAA
    if format_default:
        patron_fecha = r'\b\d{2}/\d{2}/\d{4}\b'
    else:
        patron_fecha = r'\b\d{1,2}/\d{1,2}/\d{4}\b'
    # Construir una lista con las fechas encontradas
    lista_fechas = [
        fecha
        for fila in dataset["texto"]
        for fecha in re.findall(patron_fecha, fila)
    ]
    # Convertir la lista en una Serie de pandas
    serie_fechas = pd.Series(lista_fechas)
    return serie_fechas

# Extraer las fechas válidas del DataFrame
fechas_validas = extraer_fechas(df)

# Imprimir el resultado
print("Fechas válidas encontradas:")
print(fechas_validas)
print("Total de fechas:")
print("Nota: Considerando que DD y MM pueden tener un solo dígito:")
print(extraer_fechas(df, format_default=False))
```

```python
# Resuelva en esta casilla el ejercicio 03
import pandas as pd
import re

# Guardar en la variable 'ruta' la url del dataset
ruta_csv = (
    "https://docs.google.com/spreadsheets/d/e/"
    "2PACX-1vTRfE6-FxXXP9jFLrFCuQZr6MTB5uYToBnJ"
    "QJKqjuQJfESFAsh5-fC9A14lpLfFnda3if8ioCkbhsfW/"
    "pub?gid=1040631360&single=true&output=csv"
)

# Cargar el dataset a partir de la ruta establecida
try:
    df = pd.read_csv(ruta_csv, header=None, dtype=str)
    df = df.rename(columns={0: "texto"})
    print("Lectura del dataset exitosa")
except Exception as e:
    print(f"Error al leer el dataset: {e}")

# Función para extraer URLs válidas
def extraer_urls(dataset):
    """
    Extrae las URLs válidas que comienzan con
    "http://" o "https://" de un DataFrame.

    Args:
    - dataset: Un DataFrame de pandas que contiene texto de todo tipo.

    Returns:
    - pandas.Series: Una serie que contiene las URLs válidas.
    """
    # URLs que comienzan con "http://" o "https://"
    patron_url = r'\b(?:https?://)\S+'
    # Construir una lista con las URLs encontradas
    lista_urls = [
        url
        for fila in dataset["texto"]
        for url in re.findall(patron_url, fila)
    ]
    # Convertir la lista en una Serie de pandas
    serie_urls = pd.Series(lista_urls)
    return serie_urls

# Extraer las URLs válidas del DataFrame
urls_validas = extraer_urls(df)

# Imprimir el resultado
print("URLs válidas encontradas:")
print(urls_validas)
```

Escriba un programa que lea el archivo de referencia y extraiga todas las URLs de imágenes válidas. Las URLs de imágenes válidas pueden terminar en extensiones comunes como ".jpg", ".png" o ".gif".

```python
# Resuelva en esta casilla el ejercicio 07
import pandas as pd
import re

# URL del archivo de referencia
ruta_csv = (
    "https://docs.google.com/spreadsheets/d/e/"
    "2PACX-1vTRfE6-FxXXP9jFLrFCuQZr6MTB5uYToBnJ"
    "QJKqjuQJfESFAsh5-fC9A14lpLfFnda3if8ioCkbhsfW/"
    "pub?gid=1040631360&single=true&output=csv"
)

# Leer el archivo CSV y cargarlo en un DataFrame
try:
    df = pd.read_csv(ruta_csv, header=None, dtype=str)
    df = df.rename(columns={0: "texto"})
    print("Lectura del dataset exitosa")
except Exception as e:
    print(f"Error al leer el dataset: {e}")

# Función para extraer URLs de imágenes válidas
def extraer_urls_imagenes(dataset):
    """
    Extrae las URLs de imágenes válidas del texto de un DataFrame.

    Args:
    - dataset: Un DataFrame de pandas que contiene texto de todo tipo.

    Returns:
    - pandas.Series: Una serie de pandas que contiene las URLs
    de imágenes válidas extraídas.
    """
    # Patrón de expresión regular para encontrar URLs de imágenes válidas
    patron_url_imagen = r'https?://[^\s]+\.(?:jpg|png|gif)'

    # Construir una lista con las URLs de imágenes encontradas
    lista_urls_imagenes = [
        url_imagen
        for fila in dataset["texto"]
        for url_imagen in re.findall(patron_url_imagen, fila)
    ]

    # Convertir la lista en una Serie de pandas
    serie_urls_imagenes = pd.Series(lista_urls_imagenes)
    return serie_urls_imagenes

# Extraer URLs de imágenes válidas del DataFrame
urls_imagenes_validas = extraer_urls_imagenes(df)

# Imprimir el resultado
print("URLs de imágenes válidas encontradas:")
print(urls_imagenes_validas)
```

## Ejercicio 09

Escriba un programa que lea el archivo de referencia y reporte cuántas palabras en mayúsculas contiene y cuántas veces aparece cada palabra.


```python
# Resuelva en esta casilla el ejercicio 09
import pandas as pd
import re

# Guardar en la variable 'ruta' la URL del dataset
ruta_csv = (
    "https://docs.google.com/spreadsheets/d/e/"
    "2PACX-1vTRfE6-FxXXP9jFLrFCuQZr6MTB5uYToBnJ"
    "QJKqjuQJfESFAsh5-fC9A14lpLfFnda3if8ioCkbhsfW/"
    "pub?gid=1040631360&single=true&output=csv"
)

# Cargar el dataset a partir de la ruta establecida
try:
    df = pd.read_csv(ruta_csv, header=None, dtype=str)
    df = df.rename(columns={0: "texto"})
    print("Lectura del dataset exitosa")
except Exception as e:
    print(f"Error al leer el dataset: {e}")

# Función para contar palabras en mayúsculas y reportar sus frecuencias
def contar_palabras_mayusculas(dataset):
    """
    Cuenta las palabras en mayúsculas en un DataFrame y reporta sus
    frecuencias.

    Args:
    - dataset: Un DataFrame de pandas que contiene texto de todo tipo.

    Returns:
    - pandas.DataFrame: Un DataFrame que contiene las palabras en mayúsculas
    y sus frecuencias.
    """
    # Patrón para buscar palabras en mayúsculas
    patron_mayusculas = r'\b[A-Z]+\b'
    # Construir una lista con las palabras en mayúsculas encontradas
    palabras_mayusculas = [
        palabra
        for fila in dataset["texto"]
        for palabra in re.findall(patron_mayusculas, fila)
    ]
    # Contar las frecuencias de las palabras en mayúsculas
    frecuencias = {}
    for palabra in palabras_mayusculas:
        frecuencias[palabra] = frecuencias.get(palabra, 0) + 1
    # Crear un DataFrame a partir del diccionario de frecuencias
    df_frecuencias = pd.DataFrame(
        frecuencias.items(),
        columns=['Palabra', 'Frecuencia']
    )
    return df_frecuencias

# Contar palabras en mayúsculas y sus frecuencias en el DataFrame
df_palabras_mayusculas = contar_palabras_mayusculas(df)

# Imprimir el resultado
print("Palabras en mayúsculas y sus frecuencias:")
print(df_palabras_mayusculas)
```
